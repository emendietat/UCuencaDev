Para esta sesión de ejercicios se tendrá que medir la velocidad de ejecución de un programa. Para ello, se utilizará la función `time()` el paquete `time`. El siguiente ejemplo calcula el tiempo de ejecución de los comandos ejecutados entre las variables `inicio` y `fin`. El resultado se muestra en segundos.

Escriba una función `calcular_suma_lista()` que reciba una lista como parámetro. Esta función retornará la sumatoria de todos los elementos de la lista. Este valor deberá ser calculado utilizando una variable acumuladora y un `for`. Además, genere listas con números aleatorios con los siguientes tamaños: [90, 900, 9000, 90000, 900000, 9000000, 10000000, 40000000, 60000000, 90000000]. Los elementos de estas listas tendrán un rango de 0 a 9. Calcule cuanto tiempo le toma a la función calcular un valor para todas las listas generadas. Al final muestre un gráfico de línea donde se muestre la evolución del tiempo de ejecución de la función del número de elementos de la lista.

```python
def generar_lista(longitud):
    '''Función que genera una lista de números aleatorios del 1 al 9 mediante la función
        random de numpy, donde se pide el tamaño de la lista.
    
    PARÁMETROS
    ----------
    longitud : int : tamaño de la lista que la función genera
    
    RETORNA
    -------
    lista : list : lista de números enteros
    '''

    lista = np.random.randint(0, 10, longitud)
    
    return lista


def calcular_suma_lista(lista):
    '''Función que crea una sumatoria de una lista mediante una variable acumuladora
        y un bucle for.
    
    PARÁMETROS
    ----------
    lista : list : lista de numeros del 1 al 9 de longitud determinada
    
    RETORNA
    -------
    sumatoria : int : suma total de todos los números de la lista ingresada
    '''
    
    sumatoria = 0
    for i in lista:
        sumatoria += i
        
    return sumatoria


def generar_grafico_lineas(valores_prueba, tiempos, labels):
    assert type(valores_prueba) == list, 'debe ser una lista de valores de prueba.'
    assert type(tiempos) == list, 'debe ser una lista de tiempos.'
    assert type(labels) == list, 'debe ser una lista de etiquetas.'
    assert len(tiempos) == len(labels)

    colores = ['blue', 'red', 'green', 'orange', 'brown', 'gray', 'yellow']

    # Arma el plot y define el tamaño
    fig, ax = plt.subplots(1, 1, figsize=(9,5))

    for i in range(len(tiempos)):

        assert type(tiempos[i]) == list, 'Los tiempos deben ser listas de valores de tipo float.'
        assert type(labels[i]) == str, 'Los labels deben ser valores de tipo String.'

        ax.plot(valores_prueba, tiempos[i], label = labels[i], color = colores[i])
        
    ax.set_xlabel('Numero de elementos')
    ax.set_ylabel('Tiempo de ejecucion [seg]')
    ax.legend() 

    plt.show()


# Lista de las longitudes de las listas
valores_prueba = [90, 900, 9000, 90000, 900000, 9000000, 10000000, 40000000, 60000000, 90000000]

# Lista de tiempos de demora de ejecución por cada lista de la funcion variable acumulatoria
tiempos_fun_1 = []

# Bucle que ejecuta la funcion de sumatoria con cada longitud de lista y calcula tiempo tardado
for i in valores_prueba:
    inicio = time.time()
    
    # Ejecuta la función de sumatoria a su vez generando una lista 
    calcular_suma_lista(generar_lista(i)) 
    
    fin = time.time() 
    tiempos_fun_1.append(fin - inicio) # almacena cada tiempo tardado

    
    # Muestra cada tiempo tardado en realizar cada sumatoria
    print(f"Para una lista de {i} elementos se demoró un tiempo de: {fin - inicio} s")

# Muestra el grafico de las cantidades de elementos y su respectivo tiempo
generar_grafico_lineas(valores_prueba, [tiempos_fun_1], ['Sumatoria usando un bucle for.'])
```

Si implementó correctamente el ejercicio, se debería ver una función lineal. Por lo tanto, esta función tiene una complejidad computacional de $O(n)$

## Ejercicio 2

Realice la misma implementación que el ejercicio 1, pero esta vez use arrays de numpy y la función `sum` de numpy. ¿Existe alguna diferencia de rendimiento?

```python
def calcular_suma_lista_np(lista):
    '''Función que crea una sumatoria de una lista mediante la función "sum" de Numpy.
    
    PARÁMETROS
    ----------
    lista : list : lista de numeros del 1 al 9 de longitud determinada
    
    RETORNA
    -------
    sumatoria : int : suma total de todos los números de la lista ingresada
    '''
    sumatoria = np.sum(lista, axis=0)
    
    return sumatoria



# Lista de tiempos de demora de ejecución por cada lista de la funcion de sum de Numpy
tiempos_fun_2 = []

# Bucle que ejecuta la funcion de sumatoria con cada longitud de lista y calcula tiempo tardado
for i in valores_prueba:
    inicio = time.time()
    
    # Ejecuta la función de sumatoria a su vez generando una lista
    calcular_suma_lista_np(generar_lista(i))
    
    fin = time.time()
    tiempos_fun_2.append(fin - inicio) # almacena cada tiempo tardado
    
    # Muestra cada tiempo tardado en realizar cada sumatoria
    print(f"Para una lista de {i} elementos se demoro un tiempo de: {fin - inicio} s")
        

# Muestra el grafico de las cantidades de elementos y su respectivo tiempo
generar_grafico_lineas(valores_prueba, [tiempos_fun_2], ['Sumatoria usando la funcion sum() de numpy.'])
```

Si bien la complejidad de esta función permanece lineal, su rendimiento mejoró ostensiblemente

Implemente tres funciones `crear_pares` que reciban como parámetros 2 listas. Genere listas de forma aleatoria usando los siguientes tamaños: 10, 100, 1000, 10000, 15000, 22000, 30000. Calcule el tiempo de ejecución para cada función y muestre los resultados en un gráfico de líneas donde se muestren los resultados del rendimiento de cada función.

```python
def crear_pares_bucle_anidado(lista1, lista2):
    '''
    '''
    pares = []
    for i, e1 in enumerate(lista1):
        for j, e2 in enumerate(lista2):
            if i == j:
                pares.append([e1, e2])
    
    return pares


def crear_pares_un_bucle(lista1, lista2):
    '''
    '''
    pares = []
    for i in range(len(lista1)):
        pares.append([lista1[i], lista2[i]])
    
    return pares


def crear_pares_numpy(lista1, lista2):
    '''
    '''
    pares = np.column_stack([lista1, lista2])
    
    return pares


def calcula_tiempo(f, listas):
    tiempo = 0
    inicio = time.time()
    f(listas[0], listas[1])
    fin = time.time()

    tiempo = fin - inicio
    return tiempo


valores_prueba = [10, 100, 1000, 10000, 15000, 22000, 30000]

tiempos_fun_1 = []
tiempos_fun_2 = []
tiempos_fun_3 = []

for i in valores_prueba:
    listas = [generar_lista(i), generar_lista(i)]
    tiempos_fun_1.append(calcula_tiempo(crear_pares_bucle_anidado, listas))
    tiempos_fun_2.append(calcula_tiempo(crear_pares_un_bucle, listas))
    tiempos_fun_3.append(calcula_tiempo(crear_pares_numpy, listas))


tiempos = [tiempos_fun_1, tiempos_fun_2, tiempos_fun_3]
labels = ['Crear pares usando 2 bucles anidados.', 'Crear pares usando un bucle.', 'Crear pares usando la función column_stack de numpy']

generar_grafico_lineas(valores_prueba, tiempos, labels)
```

Se evidencia que la función que crea pares usando dos bucles crece de forma exponencial. Por ende, tiene una complejidad de tipo $O(n^2)$ en comparación con las otras 2 implementaciones que tienen una complejidad lineal $O(n)$

## Ejercicio 5

Uno de los algoritmos más simples de ordenamiento es el de la Burbuja. Este algoritmo tiene una complejidad $O(n^2)$. Para este ejercicio, implemente el ordenamiento de la Burbuja como una función (puede buscar la implementación en internet y averiguar su funcionamiento). Luego, comparar su eficiencia con respecto a la función `sort` del paquete `numpy` para los siguientes tamaños de array [10, 100, 1000, 10000, 15000, 22000, 30000]. Los arrays de elementos serán generados de forma aleatoria y tendrán valores de tipo entero.

```python
def ordenar_por_metodo_burbuja(lista):
    for i in range(len(lista)):
        for j in range(len(lista) - 1):
            if lista[j] > lista[j + 1]:
                lista[j + 1], lista[j] = lista[j], lista[j + 1]
    
    return lista


def calcular_tiempo(f, lista):
    tiempo = 0
    inicio = time.time()
    f(lista)
    fin = time.time()
    tiempo = fin - inicio
    return tiempo


ordenar_por_npsort = lambda lista: np.sort(lista)

tiempos_fun_1 = []
tiempos_fun_2 = []

for i in valores_prueba:
    lista = generar_lista(i)
    tiempos_fun_1.append(calcular_tiempo(ordenar_por_metodo_burbuja, lista))
    tiempos_fun_2.append(calcular_tiempo(ordenar_por_npsort, lista))

generar_grafico_lineas(valores_prueba, [tiempos_fun_1, tiempos_fun_2], ['Ordenamiento usando el método burbuja.', 'Ordenamiento usando la función sort() de numpy.'])
```

Implemente la función `buscar_secuencial(x, num)` que reciba como parámetro una lista `x` y un número a buscar `num`. Esta función retornará el índice del elemento si lo encuentra; caso contrario, retornará -1. Genere 50 números enteros aleatorios entre 0 y 1000000 y ejecute la función `buscar_secuencial`. Por cada número generado, calcule el tiempo que le tomó a la función retornar un valor. Finalmente, construya un histograma con los valores de tiempo de ejecución.

```python
def buscar_secuencial(x, num):
    indice =  -1
    for key, value in enumerate(x):
        if num == value:
            indice = key
            break

    return indice


def generar_histograma(lista, bins):
    fig, ax = plt.subplots(1, 1, figsize=(15,5))
    ax.hist(lista, bins = bins)
    ax.set_xlabel('Tiempo de ejecución [Segundos]')
    plt.show()


generar_lista_random = lambda longitud, inicio, fin: np.random.randint(inicio, fin, longitud)

lista = generar_lista_random(10000000, 0, 1000000)
numeros = generar_lista_random(50, 0, 1000000)

tiempos_fun_1 = []

for i in numeros:
    inicio = time.time()
    indice = buscar_secuencial(lista, i)
    fin = time.time()
    tiempos_fun_1 .append(fin-inicio)

generar_histograma(tiempos_fun_1 , 15)
```
