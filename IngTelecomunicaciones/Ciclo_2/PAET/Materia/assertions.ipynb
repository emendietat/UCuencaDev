## Assertions (Afirmaciones)

Chequeos de 'sanidad' en código fuente:

* Documentar
* Depurar
* Probar

Ayudan a generar código fuente robusto y confiable.

```python
numero = 42
assert numero > 0, f'Se esperaba un número positivo, se tiene: {numero}'

numero = -42
assert numero > 0, f'Se esperaba un número positivo, se tiene: {numero}'
```

Sintaxis:
`assert condición[, mensaje]`

```python
import math

def calcular_area_circulo(radio):
  assert isinstance(radio, int) and radio > 0, f'Se esperaba un número entero y positivo como radio, se tiene: {radio}'
  return math.pi * radio ** 2

print(calcular_area_circulo(10))
print(calcular_area_circulo(10.0))
print(calcular_area_circulo(-10))
```

Las assertions son útiles para:

* Verificar si el código es correcto (depuración).
* Condición de assertion es verdadera a menos que ocurra una situación no deseada.
* Si condición es falsa, se interrumpe ejecución y se muestra mensaje.
* Chequeo de precondiciones (al inicio de una función).
* Chequeo de postcondiciones (antes de retornar de una función).
* Documentar código.
* Alertas de errores al escribir código.

```python
def calcular_area_rectangulo(coordenadas):
  #coordenadas almacena las coordenadas de las esquinas inferior izquierda y superior derecha del rectángulo
  assert isinstance(coordenadas, list), 'Las coordenadas deben especificarse en una lista'
  assert len(coordenadas) == 4, 'Se esperaba 4 coordenadas'

  x0, y0, x1, y1 = coordenadas

  assert x0 < x1, 'Coordenadas en X no válidas'
  assert y0 < y1, 'Coordenadas en Y no válidas'

  dx = x1 - x0
  dy = y1 - y0
  area = dx * dy

  assert 0 < area <= 50, 'El área no debe superar 50'
  return area

print(calcular_area_rectangulo([1, 1, 5, 10]))
```
