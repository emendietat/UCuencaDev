## Ejemplo: Factorial

```python
numero = 6                          #1 operación
respaldo = numero                   #1 operación
resultado = 1                       #1 operación

while numero > 1:                   #1 operación (se repite n veces: n = numero)
  resultado *= numero               #2 operaciones (se repiten n veces: n = numero)
  numero -= 1                       #2 operaciones (se repite n veces: n = numero)

print(f'{respaldo}! = {resultado}') #1 operación
```

Complejidad: $1 + 1 + 1 + 1*n + 2*n + 2*n + 1 = 4 + 5n$

## Creando listas de pares: Implementación 1 (complejidad cuadrática: $n^2$)

```python
lista1 = [100, 200, 300]
lista2 = [101, 201, 301]
resultado = []

for i, el1 in enumerate(lista1):

  for j, el2 in enumerate(lista2):

    if i == j:
      resultado += [[el1, el2]]

print(resultado)
```

## Creando listas de pares: Implementación 2 (complejidad lineal: $n$)

```python
lista1 = [100, 200, 300]
lista2 = [101, 201, 301]
resultado = []

for i in range(len(lista1)):
  resultado += [[lista1[i], lista2[i]]]

print(resultado)
```

## Complejidad de la Búsqueda Secuencial

```python
import numpy as np

arreglo = np.random.randint(1, 1000000, 1000000)
valor_buscar = np.random.randint(1, 1000000);
posicion = -1

for i, el in enumerate(arreglo):

  if el == valor_buscar:
    posicion = i

if posicion != -1:
  print(f'El número {valor_buscar} se encontró en la posición {posicion}')
else:
  print(f'El número {valor_buscar} no se encontró en el arreglo')
```

* **Mejor caso:** valor buscado está al inicio del arreglo
* **Peor caso:** valor buscado no está en el arreglo

Al analizar la complejidad de un script se considera siempre el peor caso.

## Notación Big O

Forma compacta de expresar la complejidad de un algoritmo

* **Factorial:** $O(n)$ (lineal)
* **Listas de Pares (Implementación 1):** $O(n^2)$ (cuadrática)
* **Listas de Pares (Implementación 2):** $O(n)$ (lineal)
* **Búsqueda Lineal:** $O(n)$ (lineal)
