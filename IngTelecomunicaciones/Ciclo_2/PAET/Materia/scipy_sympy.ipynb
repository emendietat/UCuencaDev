## Integración Numérica

Determinar $\int_{-2}^{2}{e^{cos(-2x\pi)}+3.2}$

```python
import math
import scipy.integrate as spint

def f(x):
  return math.exp(math.cos(-2 * x * math.pi)) + 3.2

resultado, error = spint.quad(f, -2, 2)
print('Resultado: {:f}, Error: +/-{:g}'.format(resultado, error))
```

## Raíces de una función

Encontrar las raíces de la función $f(x) = x^3 - 2x^2$

```python
import scipy.optimize as spopt

f = lambda x: x**3 - 2 * x**2
x = spopt.bisect(f, 1.5, 3, xtol=1e-6)
print('Corte de f(x) con el eje horizontal: ', x)
```

## SimPy

Cálculos algebraicos usando símbolos (variables)

```python
import sympy as sp

x, y = sp.symbols('x y')
z = x + 2*y

print('z = ', z)
print(z + 1)
print(x * z)
print(x + y + x + 10 * y)
```

## Sustitución de variables por valores

```python
x, y = sp.symbols('x y')
z = x + 2*y

print('z = ', z)
print(z.subs(x, 10))
print(z.subs({x: 10, y: 3}))
```

## Números racionales

```python
num_rac = sp.Rational(1, 10)

print(num_rac)
print(num_rac + num_rac)
print(float(num_rac + num_rac))
```

## Derivada de una función

Encontrar derivadas de $f(x) = sen(x)$ y de $f(x, y) = x^9 + 10x^2 + 3x + 4y + 10$

```python
import sympy as sp

x, y = sp.symbols('x y')

print('Derivada de f(x) = sen(x): ', sp.diff(sp.sin(x), x))
df_dx = sp.diff(x**9 + 10*x**2 + 3*x + 4*y + 10, x)
print('df(x, y)/dx = ', df_dx)
print('df(1.5, y)/dx = ', df_dx.subs(x, 1.5))
df_dy = sp.diff(x**9 + 10*x**2 + 3*x + 4*y + 10, y)
print('df(x, y)/dy = ', df_dy)
print('3ra. derivada de f(x, y) con respecto a x:', sp.diff(x**9 + 10*x**2 + 3*x + 4*y + 10, x, x, x))
```

## Sistemas de Ecuaciones Lineales

Resolver el siguiente sistema de ecuaciones lineales:

$3x + 7y = 12z$

$4x - 2y = 5z$

```python
x, y, z = sp.symbols('x y z')
ecuacion1 = sp.Eq(3*x + 7*y, 12*z)
ecuacion2 = sp.Eq(4*x - 2*y, 5*z)
solucion = sp.solve((ecuacion1, ecuacion2), x, y)
print('Solución: ', solucion)

print('\nPara z = 3.3:')
print('x = ', solucion[x].subs({z: 3.3}))
print('y = ', solucion[y].subs({z: 3.3}))
```
