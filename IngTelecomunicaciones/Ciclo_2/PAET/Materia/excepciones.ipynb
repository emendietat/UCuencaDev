## Manejando excepciones: `try`y `except`

En Python, las instrucciones `try` y `except` se usan para atrapar y manejar excepciones.

image.png

* Las instrucciones dentro del bloque `try` se ejecutan normalmente.
* Si una de las instrucciones dentro de `try` genera una excepción, la ejecución del bloque `try` se detiene, y esa excepción puede manejarse dentro de un bloque `except`.

```python
try:
  dividendo = int(input('Dividendo: '))
  divisor = int(input('Divisor: '))
  print(dividendo / divisor)
  print('El bloque try continúa su ejecución')
except ZeroDivisionError as mensaje_error:
  print('El divisor no puede ser cero')
except ValueError as mensaje_error:
  print('Tanto el dividendo como el divisor deben ser enteros')

print('El script continúa su ejecución')
```

```python
import pandas as pd

try:
  llamadas = pd.read_csv('/content/phone_data.csv', sep=',', index_col=0, parse_dates=[1])
  print('El bloque try continúa su ejecución')
except FileNotFoundError as mensaje_error:
  print(mensaje_error)

print('El script continúa su ejecución')
```

## El bloque `else`

image.png

```python
try:
  print(10 / 2)
  print('El bloque try continúa su ejecución')
except ZeroDivisionError as mensaje_error:
  print(mensaje_error)
else:
  print('No se generó ninguna excepción')
```

## El bloque `finally`

image.png

```python
try:
  print(10 / 5)
  print('El bloque try continúa su ejecución')
except ZeroDivisionError as mensaje_error:
  print(mensaje_error)
else:
  print('No se generó ninguna excepción')
finally:
  print('Las instrucciones del bloque finally se ejecutan haya o no excepciones')
```
